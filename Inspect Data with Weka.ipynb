## Check the number of NAN values in each dataset
Note NAN can either be:
- missing values
- varying column length

### NAN in UCR

```python
datasetPath = "C:\\Users\\Paws\\Desktop\\MasterProject\\ExtraDatasets\\UCRArchive_2018"

keys = []
values =[]
for filename in os.listdir(datasetPath):
    if filename == "Missing_value_and_variable_length_datasets_adjusted":
        continue
    
    train, train_labels, test, test_labels = load_UCR_dataset(datasetPath, filename)
    #number of missing values in each dataset
    num_NAN = np.count_nonzero(np.isnan(train))
    if num_NAN != 0:
        index_NAN = np.argwhere(np.isnan(train))
        keys.append(filename)
        values.append(index_NAN)
        print(filename, num_NAN)
        
UCR_NAN_dict = dict(zip(keys,values))


#Pickle the dictionary containing indexes of NAN values
pickle.dump(UCR_NAN_dict, open('../ExtraDatasets/NAN_info/UCR_NAN_dict.pth','wb'))


#To unpickle, use the following code format
#pickle.load(open(file_path,'rb'))
#open(file_path,'rb').close()
# 
# UCR_NAN_dict = pickle.load(open('../ExtraDatasets/NAN_info/UCR_NAN_dict.pth','rb'))
# open('../ExtraDatasets/NAN_info/UCR_NAN_dict.pth','rb').close()
```

### NAN in UEA

```python
datasetPath = "C:\\Users\\Paws\\Desktop\\MasterProject\\ExtraDatasets\\Multivariate_arff"

# keys = []
# values =[]
for i, filename in enumerate(os.listdir(datasetPath)):
    if filename == "DataDimensions.csv" or filename == "Descriptions" or filename == "Images":
        continue
    elif i < 17:
        continue
        
    train, train_labels, test, test_labels = load_UEA_dataset(datasetPath, filename)
    #number of missing values in each dataset
    num_NAN = np.count_nonzero(np.isnan(train))
    if num_NAN != 0:
        index_NAN = np.argwhere(np.isnan(train))
        keys.append(filename)
        values.append(index_NAN)
        print(filename, num_NAN)

UEA_NAN_dict = dict(zip(keys,values))


#Pickle the dictionary containing indexes of NAN values
pickle.dump(UEA_NAN_dict, open('../ExtraDatasets/NAN_info/UEA_NAN_dict.pth','wb'))


#To unpickle, use the following code format
#pickle.load(open(file_path,'rb'))
#open(file_path,'rb').close()
# 
# UCR_NAN_dict = pickle.load(open('../ExtraDatasets/NAN_info/UEA_NAN_dict.pth','rb'))
# open('../ExtraDatasets/NAN_info/UEA_NAN_dict.pth','rb').close()
```
